A Poupança é um investimento muito popular no nosso país.
E comumente conhecida como um bom lugar para deixar o dinheiro guardado.

Mas isto é o que ainda é aceito pela maioria - maioria esta com um menor conhecimento e interação sobre os assuntos de finanças.

# Por que a poupança tem baixa rentabilidade?
poupançaEmBaixa.jpg

Ela tem como referência a nossa básica de juros - **lady Selic**

Mas há 2 casos:

- Quando a SELIC menor ou igual a 8,5% ( sendo esta uma taxa anual)

Onde a rentabilidade é de  70% da SELIC.

Isto significa uma taxa de rentabilidade de 0,5 % ao mês.

    





###Exemplo de uma rentabilidade no período de 1 ano

com um capital de R$ 1000.00 (mil reais).

**Com uma matemática bem simplista.**





rentabilidade = capital * (0,7 * SELIC)

 $= 1000 * (0,7 * 0,085)$

 $= 1000 * 0,059$

$= 59$




- Quando a SELIC está maior que 8,5 %...
    
   A sua rentabilidade vai ter uma taxa fixa de 0,5% ao mês na poupança.
    
   Mais  a Taxa Referencial. Uma taxa tão baixa que é pouco conhecida.




    



    
    

rentabilidade = capital * 0,5%

 $= 1000 * 0,005$

$= 50$
