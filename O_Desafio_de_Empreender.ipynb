Após feita esta limpeza inicial nos 2 dataframes vamos preparar o nosso intruso. para que ele possa se juntar aos aspiras.

```python
intruso.drop(columns=['anos_negócio','funcionários_noComeço',	'funcionários_hoje', 'fontes_de_renda','Dificuldade Evolutiva','visão'], inplace=True)

# trocando colunas
def swap_columns(df, col1, col2):
    col = list(df.columns)
    x, y = col.index(col1), col.index(col2)
    col[y], col[x] = col[x], col[y]
    df = df[col]
    return df
```

```python
# adicionar coluna faltante
intruso.insert(3, 'empreender', 'Não')

intruso = swap_columns(intruso, 'Gestão', 'networking')
intruso = swap_columns(intruso, 'preço_de_trabalho', 'evoluir_mais')

# renomeando colunas
intruso.rename(columns={'Plano':'nível_de_planejamento','se_preparou': 'se_prepararia',
                        'momento_hoje': 'o_que_faria_hoje','evoluir_mais':'sempre_em_evolução',
                        'aposentar':'aposentar_empreendendo'}, inplace=True)

# adicionando intruso á última linha
nova_linha = []
nova_linha.append(intruso)
aspiras = aspiras.append(nova_linha)

# atualizando índice do dataframe
aspiras.reset_index(drop=True, inplace=True)

# formatando valores ausentes
aspiras.fillna(0, inplace=True)

locals = [4,7,12]
update = ['Não penso em abrir um negócio.', 'Sim. Estudando', 4]

for i in range(3):
  aspiras.iloc[8,locals[i]] = update[i]
```

# Auto avaliação

Com algumas das questões formadas sob a escala de Likert temos uma visão amplificada das dificuldades e facilidades para este público.

```python
# criando variável
avalia = ceos[['Dificuldade Evolutiva','Medo',
               'Comunicação', 'Sinceridade', 'Gestão', 'Plano']]

# igualando tipos de valores
avalia[['Plano','Medo']] = avalia[['Plano','Medo']].astype(int)

# criando estrutura de dicionário
frequencia = {}

# adiciconando valores a partir da variável
for i in avalia.columns:
  frequencia[i] = avalia[i].value_counts()

# formatando conjunto
plotar = pd.DataFrame(frequencia)
plotar = plotar.transpose().fillna(0)
```

```python
shades = ['gray', 'red','gold','#2cbd1c','#1c85bd']

plotar.plot(kind='barh', color= shades)

leg = ['Discordo Plenamente', 'Discordo', 'Neutro', 'Concordo','Concordo Plenamente']
plt.legend(leg, bbox_to_anchor=(0.5, -.2), loc='center',
           fontsize= 8,
           ncol= 5)

ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)
ax.spines.bottom.set_position(('outward', 10))
```

* A maioria dos CEOs afirmam que têm um plano, mas é bem dividido o fato de estarem sempre de acordo com o mesmo.


* Quanto a confiança na gestão a maioria também apresentou em nível médio. Apenas um demonstrando extrema confiança nesta parte, e um volme mínimo de discordância.
* Todos disseram ser sinceros em seus relacionamentos de trabalho. Seja com colaboradores, clientes ou fornecedores.
* A habilidade de comunicação apresenta 2 unidades próximas da discordância. E a maioria sente esta confiança.
* O medo representa "medo ou incertezas".

E é um ponto muito interessante.

Muitos confirmaram o reconhecimento deste sentimento.

É o que é! E não que seja ruim.

Este pode impulsionar para ambos os lados: evolução ou estagnação.

E o Empreendedorismo é desafio. O desafio demanda coragem, pois significa atravessar pelo medo.

E a dificuldade evolutiva se assenta sobre a maioria e é reconhecida por eles.

```python
evalu = aspiras[['Segue o Plano', 'Medo', 'Comunicação',
                 'Sinceridade','Gestão','Plano']]

evalu.astype(int)

freq = {}

for v in evalu.columns:
  freq[v] = evalu[v].value_counts()

asps_agree = pd.DataFrame(freq)
asps_agree = asps_agree.transpose().fillna(0)


shades = ['gray', 'red','gold','#2cbd1c','#1c85bd']

asps_agree.plot(kind='barh', color=shades)


legs = ['Discordo Plenamente', 'Discordo', 'Neutro', 'Concordo','Concordo Plenamente']
plt.legend(leg, bbox_to_anchor=(0.5, -0.15), loc='center', ncol=5)

ax = plt.gca()
ax.spines[['left','top','right']].set_visible(False)
```
